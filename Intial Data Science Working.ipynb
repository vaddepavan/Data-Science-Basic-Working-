23. Use the codes for each country to merge the files so that you can convert the salary data to Australian Dollars (AUD). Print out the maximum and median salary in AUD. Hint: think about what data type you have.

```python
# Your code
# merging the both data files to obtain the desired salary in AUD
df7 = df.iloc[:,[-3,-2]] # only fetching the compenssation amount and compensation currency from one dataframe
df8 = df1.iloc[:,[0,2]] # only fetching the origin country  and exchange rate from the another data frame  
df9 = pd.merge(df7, df8, left_on="CompensationCurrency", right_on="originCountry") # merging the sliced data frames and converting the currency into AUD
df9['CompensationAmountAUS'] = df9.CompensationAmount*df9.exchangeRateAUS
print("Maximum Salary in AUD = ",df9.CompensationAmountAUS.max())#taking the mean and median of the salaries
print("Median Salary in AUD = ",df9.CompensationAmountAUS.median())
df9.CompensationAmountAUS.describe() # presenting the statistics  
```
